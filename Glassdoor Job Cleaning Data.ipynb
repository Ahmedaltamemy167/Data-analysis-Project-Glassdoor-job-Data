```python
df= df[df["Salary Estimate"] != '-1']
```

```python
df["Salary Estimate"].value_counts()
```

```python
salary = df["Salary Estimate"].apply(lambda x : x.split('(')[0])
salary
```

```python
salary = salary.str.replace('$','')
salary = salary.str.replace('K','')
salary
```

```python
salary.unique()
```

```python
salary = salary.str.replace('Employer Provided Salary:',"")
salary = salary.str.replace('Per Hour',"")
```

```python
salary.unique()
```

```python
df["min_salary"]= salary.str.split('-').str[0]
df["min_salary"] 
```

```python
df["max_salary"]=salary.str.split('-').str[1]
df["max_salary"]
```

```python
df.head()
```

```python
df["min_salary"]= df["min_salary"].astype(int)
```

```python
df["max_salary"] = df["max_salary"].astype(int)
```

```python
df.dtypes
```

```python
df["avg_salary"] = (df["min_salary"]+ df["min_salary"])/2
```
